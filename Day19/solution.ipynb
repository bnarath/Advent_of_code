```python
def parse(file):
    rule_dict = {}
    string_list = []
    with open(file, 'r') as f:
        for string in f.readlines():
            if re.match(r'^\d', string):
                index_rule_string = string.split(':')
                rule_combo = index_rule_string[1].strip().split('|')
                try:
                    rule_dict[int(index_rule_string[0])]=[[int(rule) for rule in rules.split()] for rules in rule_combo]
                except:
                    rule_dict[int(index_rule_string[0])] = index_rule_string[1].strip().strip("\"")
            elif string.strip():
                string_list.append(string.strip())
    return rule_dict, string_list
```

```python
rule_dict, string_list = parse('input.txt')
```

```python
def cross_concat(a, b):
    if not a and b:
        return b
    elif a and not b:
        return a
    else:
        List = []
        for i in a:
            for j in b:
                List.append(i+j)
        return List
            
def findtherules(id, rule_dict, string_list,part=1):
    if type(rule_dict[id])==str:
        return [rule_dict[id]]
    else:
        rule_list = []
        for each in rule_dict[id]:
            rules = []
            for seq in each:
                rules = cross_concat(rules,findtherules(seq, rule_dict, string_list, part=1))
            rule_list = rule_list + rules
        if (part==2) and (id in [8, 11]):
            r = max([len(item) for item in findtherules(id, rule_dict, string_list, part=1)])
            max_length_string = max([len(item) for item in string_list])
            iteratn = max_length_string//r
            
            
        return rule_list

def part1(rule_dict, string_list):
    rule_list = findtherules(0, rule_dict, string_list,part=1)
    match = [string for string in string_list if string in rule_list]
    return len(match)
```

```python
def part2(rule_dict, string_list):
    r1 = findtherules(8, rule_dict, string_list,part=1)
    r2 = findtherules(11, rule_dict, string_list,part=1)
    max_length_string = max([len(item) for item in string_list]) #80
    rule_list = findtherules(0, rule_dict, string_list,part=1)
    match = [string for string in string_list if string in rule_list]
    match_with_r1_r2_dict = {}#In the rule list
    match_with_r1_r2 = []#In the string list matching r1, r2 but not in match
    for item in string_list:
        for rule in r1+r2:
            if (re.search(rule, item) and (item not in match)):
                match_with_r1_r2.append(item)
        
    for item in rule_list:
        r1s = re.search(r1, item)
        r2s = re.search(r2, item)
        if r1s:
            match_with_r1_r2_dict[item] = match_with_r1_r2_dict.get(item, []) + r1s.span()
        if r2s:
            match_with_r1_r2_dict[item] = match_with_r1_r2_dict.get(item, []) + r2s.span()    
    return match_with_r1_r2_dict, match_with_r1_r2
```
